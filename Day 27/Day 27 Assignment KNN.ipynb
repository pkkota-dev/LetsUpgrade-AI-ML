```python
import pandas as pd


df = pd.read_csv('train.csv')
df.head(5)

df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df['Sex'] = encoder.fit_transform(df['Sex'])
df['Embarked'] = encoder.fit_transform(df['Embarked'])

df=df.drop(['Cabin','PassengerId','Name','Ticket'],axis=1)
df.head()
```

```python
df = df[['Survived','Pclass','Sex','SibSp','Parch','Embarked','Age']]
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
features = ['Survived','Pclass','Sex','SibSp','Parch','Embarked']
k = []
acc = []
for i in features:
    x = df.drop(i,axis = 1)
    y = df[i]
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
    print(x_test.shape[0])
    for j in  range(1,x_test.shape[0]+1):
        knn = KNeighborsClassifier(n_neighbors=j)
        knn.fit(x_train,y_train)
        y_pred = knn.predict(x_test)
        con = confusion_matrix(y_pred,y_test)
        accuracy = accuracy_score(y_pred,y_test)
        k.append(j)
        acc.append(accuracy)
        print(i,j)
        print(con)
        print('------------------------------------------------------------------------------------')
    plt.plot(k,acc,color ='red')
    plt.title(i)
    print(acc.index(max(acc))+1)
    print(max(acc))
    plt.show()
```
