#Statistical Tests (Mann-Whitney)

```python
from sklearn import preprocessing
from scipy.stats import mannwhitneyu
label_encoder = preprocessing.LabelBinarizer()
dataset['Attrition'] = label_encoder.fit_transform(dataset['Attrition'])

print(dataset.Attrition)
```

#Attrition Vs Distance From home
<p>
H0: There is no significant differences in the Distance From Home between attrition
    </p>
    <p>
Ha: There is significant differences in the Distance From Home between attrition
    </p>

```python

stats,p = mannwhitneyu(dataset.Attrition, dataset.DistanceFromHome)
print(stats,p)
```

Conclustion: As p value is <0.05 Null hypothesis is rejected and Alternate Hypothesis is accepted

#Attrition Vs Montly Income

<p>H0: There is no significant differences in the Monthly income between attrition </p>
<p>Ha: There is significant differences in the Monthly income between attrition </p>

```python
stats,p = mannwhitneyu(dataset.Attrition,dataset.MonthlyIncome)
print(stats,p)
```

Conclustion: As p <0.05 Null Hypothesis is rejected and alternate hypothesis is accepted

#Attrition Vs YearsWithCurrManager
<p>H0: There is no significant differences in the years with current manager between attrition </p>

<p>Ha: There is significant differences in the years with current manager between attrition</p>

```python
stats,p = mannwhitneyu(dataset.Attrition,dataset.YearsWithCurrManager)
print(stats,p)
```

Conclustion: As p <0.05 Null Hypothesis is rejected and alternate hypothesis is accepted

#Chi-Square test

```python
# 1)BusinessTravel
#H0 : There is no dependency between BusinessTravel and Attrition rate.
#H1 : There is a dependency between BusinessTravel and Attrition rate

from scipy.stats import chi2_contingency
chitable = pd.crosstab(dataset.Attrition,dataset.BusinessTravel)
stats,p,dof,expeted = chi2_contingency(chitable)
print(stats,p)
```

#p<0.05 reject null hypothesis distance is not creating major impact in attrition 

```python
#Martial Status

#H0 : There is no dependency between Marital Status and Attrition rate.
#H1 : There is a dependency between Marital status and Attrition rate.

chitable = pd.crosstab(dataset.Attrition,dataset.MaritalStatus)
stats,p,dof,expeted = chi2_contingency(chitable)
print(stats,p)
```

#p<0.05 reject null hypothesis There is a dependency between martial status and attriion rate
