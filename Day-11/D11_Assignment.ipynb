```python
#Corelation between attrition and age
''' 
H0: There is no relation between attrition and age
H1: Therr is a relation between attrition and age

conclustion: 
r= -0.159   as r< 0 we can say attrition is -ve ly co-related to age
p = 1.9  as p>= 0.05 we are accepting the null hypothesis
'''
from scipy.stats import pearsonr


r,p = pearsonr(dataset.Attrition, dataset.Age)
print(r, p)

```

```python
#Corelation between attrition and DistanceFrom Home
''' 
H0: There is no relation between attrition and distance from home
H1: Therr is a relation between attrition and distance from home

conclustion: 
r= -0.00973   as r< 0 we can say attrition is -ve ly co-related to distance from home
p = 0.5  as p>= 0.05 we are accepting the null hypothesis
'''

r,p = pearsonr(dataset.Attrition, dataset.DistanceFromHome)
print(r, p)
```
